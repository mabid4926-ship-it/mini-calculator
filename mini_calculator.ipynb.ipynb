```python
#take two numbers and an arthimatic operator from an user and print
#result of corresponding operation
#(+,-.*,/)
num1=int(input("Enter a First number"))
num2=int(input("Enter a second  number"))
operator=input("enter operator")
if operator=="+":
        print("result",num1+num2) 
elif operator=="-" :
        print("result",num1-num2)
elif operator =="*":
        print("result",num1*num2)
elif operator  =="/" :
        print("result",num1/num2)
else:
   print("invalid operation") 
               
               
```
